#### Python 쯘st치ticamente tipado? 游땸游땸

Bueno... [m치s o menos][chuleta]. Realmente Python permite a침adir **anotaciones de tipo**. 쯈u칠 significa esto? Pues que **sin una aplicaci칩n externa** (como un plug-in de un IDE) **no es posible aprovechar esta opci칩n**, pues el int칠rprete de Python no lo comprueba. Esta aplicaci칩n externa se ejecuta antes que el int칠rprete, y para la ejecuci칩n si encuentra alg칰n conflicto de tipos, como ocurre en los lenguajes est치ticamente tipados.

Esta posibilidad **se introduce en [Python 3.5][pep484]** y se mejora en las siguientes versiones. Usaremos el [m칩dulo `typing`][typing] para ello.

[why]: https://wiki.python.org/moin/Why%20is%20Python%20a%20dynamic%20language%20and%20also%20a%20strongly%20typed%20language

````python
from datetime import datetime as dtime

class Persona:
    duracion_anyo: int = 365
    
    def __init__(self, nombre: str, altura: float, fecha_de_nacimiento: str):
        self.nombre = nombre
        self.altura = altura
        self.fecha_de_nacimiento = dtime.strptime(fecha_de_nacimiento, "%Y-%m-%d")
        
    def wave(self):
        diferencia = dtime.now() - self.fecha_de_nacimiento
        edad_actual = diferencia.days // Persona.duracion_anyo
        
        print(f"Mi nombre es {self.nombre}, tengo {edad_actual} a침os y mido {self.altura} cms.")
````

````python
# Necesitamos insertar la fecha en el formato "YYYY-MM-DD" ! 
Quino = Persona("Quino", 181, "1996-11-11")

Quino.wave()
````

### Decoradores

Los **decoradores** son algo muy interesante, algo 칰nico de Python. Es algo as칤 **como las macros de Excel o Rust**: permiten **aplicar la funcionalidad de una clase o una funci칩n** a uno o varios bloques del c칩digo, integr치ndose en la sintaxis del lenguaje y **permitiendo una mejor lectura**. 

Podemos entender los decoradores como <u>funciones de alto nivel</u>, pues toman como argumento una funci칩n y devuelven otra funci칩n. 칄sto lo veremos m치s adelante en el bloque funcional, pero quedate con ese concepto.

````python
def agregar_barritas(funcion_con_texto):
    """Agrega barritas arriba y abajo de la funcion"""
    def envoltorio():
        """Aplica las barritas al texto"""
        print("==================")
        funcion_con_texto()
        print("==================")
        
    return envoltorio

def illo_que_ase():
    """Una funcion que produce texto en la consola"""
    print("Illo que ase")

con_barritas_todo_es_mejor = agregar_barritas(illo_que_ase)

illo_que_ase()
print("\n")
con_barritas_todo_es_mejor()
````

Lo realmente 칰nico de los decoradores es su ***az칰car sint치ctico***. 춰Puedes declarar el decorador sobre la funcion que escribe texto con una simple **`@`**!

````python
@agregar_barritas
def illo_que_ase():
    """Una funcion que produce texto en la consola"""
    print("Illo que ase")
    
illo_que_ase()
````

Los decoradores se usan mucho en paquetes de computaci칩n num칠rica. Por ejemplo, [`numba`][numba] es una libreria que permite pre-compilar c칩digo Python a C/C++, y puede mejorar mucho c칩digo orientado a computaci칩n num칠rica. En este caso, usa decoradores `@jit` entre otros.

#### Generadores

Algo fundamental para lenguajes funcionales como Haskell, son estructuras "pausadas", es decir, en las que **todos sus valores no son calculados al crear la estructura**. El ejemplo m치s claro est치 entre `list` y `range`:



````python
lista = [0,1,2,3,4]
rango = range(0,5)

print("{} y {}".format(lista, rango))
````

Los generadores **se definen sobre estructuras iterables**, es decir, que se puedan recorrer. En cierto modo, cuando creas un iterador sobre una lista, obtienes un objeto `Iterador` sobre el que puedes usar una funci칩n **`next()`** para obtener el siguiente objeto en la estructura. Cuando la estructura llega a su fin, lanza un error `StopIteration` si ejecutas `next()` sobre el iterador ya vac칤o.

Hay dos formas de escribir generadores:

+ Si queremos crear generadores similares a listas, podemos usar una notaci칩n similar a la de por comprensi칩n:

````python
generador2 = (x ** 2 for x in [1,2,3])

print("{} {}".format(generador2, list(generador2)))
````

+ Usando la palabra `yield`. De esta forma, podemos usarlo en cualquier tipo de estructura de datos iterable:

````python
def obtener_proximo_elemento(lista):
    for indice, valor in enumerate(lista):
        yield (indice, valor)
        
iterador = obtener_proximo_elemento([1,2,3,4,5,6])
iterador
````

````python
print(f"{next(iterador)}, {next(iterador)}")

for i, v in iterador:
    print("Indice {} valor {}".format(i,v))
    
next(iterador)  # error!
````

### (**Opcional**) Documentaci칩n

*Si quieres ir r치pido, programa solo. Si quieres llegar lejos, programa con los dem치s*

Tanto si es para desarrollar herramientas m치s complejas por ti mismo como si est치s participando en un proyecto con otros desarrolladores, documentar adecuadamente es esencial. Python tiene su propia gu칤a de estilo para la documentaci칩n, **basada en reStructuredText y condensada en el [PEP287][pep287] y [PEP257][257]**. En Python, los llamamos ***docstrings***.

De manera simple, reStructuredText es un lenguaje de texto enriquecido que compila a HTML. Ve치moslo:

[pep287]: https://www.python.org/dev/peps/pep-0287/
[257]: https://www.python.org/dev/peps/pep-0257/

````python
def saludar(nombre=None, entusiasmadamente=False):
        """Saluda a una persona, con fuerza o no.

        Parametros
        ----------
        nombre : str, obligatorio
            El nombre de la persona a saludar
        entusiasmadamente: bool, opcional
            Elige si saludar con fuerza o no

        Lanza
        ------
        TypeError
            Si el nombre es None
        """

        if nombre is None:
            raise TypeError("El nombre es None")

        exclamacion = ["", "!"][entusiasmadamente]  # equivalente a un operador ternario ?: (no recomendable! haz un 'if-else'. Esto no es tan legible)
            
        print("Hola Don Pepito! Hola Don {}{}".format(nombre, exclamacion))
        
saludar("Jose", True)
saludar("Jose", False)
````

> F칤jate! Una de las mejores maneras de documentar c칩digo en Python es **asignar correctamente los nombres** a funciones y variables temporales. Ten en cuenta el estilo de Python: `estoNoEstaFino`, `esto_si_es_mejor`

El **lenguaje funcional** permite una **abstracci칩n** a la hora de **modificar estructuras de datos**. En vez de pensar c칩mo cambiar una estructura, como hacemos con un bucle `for`, **pasamos a pensar qu칠 queremos cambiar**. Esto nos ahorra, por ejemplo, errores de acceso a memoria. Adem치s, no tenemos que preocuparnos por iterar sobre la estructura (no usaremos m치s `for`, `while`, ... )

Vamos a dar **4 funciones fundamentales de los lenguajes funcionales**:

+ `map` permite <u>aplicar</u> a cada elemento de una estructura, <u>una funci칩n</u>. Por ejemplo, podemos sumar 1 a todos los elementos de una lista:


````python
list1 = [1,2,3]

def mas1(valor):
    return valor + 1

list(map(mas1, list1))
````

+ `filter` permite <u>seleccionar</u> los elementos de una estructura que cumplan la funci칩n usada. Por ejemplo, podemos escoger solo los numeros pares de una lista:

````python
def esPar(valor):
    return valor % 2 == 0

list(filter(esPar, list1))
````

+ `reduce` permite <u>reorganizar</u> una estructura A para obtener otra estructura B. Normalmente, se usa para condensar estructuras complejas y conseguir estructuras mas simples, con valores acumulados. Por ejemplo, podemos sumar todos los elementos de una lista para obtener un valor total (una estructura m치s simple):

````python
from functools import reduce

def sumar(valor1, valor2):
    # ojito! Ten en cuenta que, para 'reduce' el valor se acumula en el segundo (2do) valor
    return valor1 + valor2

reduce(sumar, list1)
````

+ `zip` permite <u>agrupar</u> varias estructuras $A_x$ en una sola estructura $B$. La estructura $B$ est치 <u>formada por tuplas de los elementos de las estructuras $A_x$</u>. Por ejemplo, podemos agrupar los nombres y alturas de personas en una sola lista:

````python
nombres = ["Quino", "Andres", "Fernando"]
altura = [181, 178, 180]

list(zip(nombres, altura))
````

### (**Opcional**) Funciones parciales

La idea principal es que **toda funci칩n $A$ con $N$ par치metros devuelve otra funci칩n $B$ con $N-1$ par치metros**, donde $B$ **tiene un estado interno diferente a $A$**. 칄sto quiere decir que, al igual que las clases, las funciones pueden mantener un estado interno (como la referencia *self* en las clases!). Un ejemplo cl치sico es el c치lculo del [n칰mero n de *fibonacci*][fib]:

$f_{0}=0\,$

$f_{1}=1\,$

$f_n = f_{n-1} + f_{n-2}$

[fib]: http://www.pythondiario.com/2018/08/sucesion-de-fibonacci-con-python.html

````python
def configurar_fib():
    a, b = 0, 1  # mantenemos un estado interno en la funci칩n. Esto se conoce como "memoization"
    def fib(n):
        if n == 0:
            return a
        elif n == 1:
            return b
        else:
            return fib(n-1) + fib(n-2)
    return fib
        
fib = configurar_fib()

[fib(x) for x in range(13)]
````

Existen dos maneras de crear funciones parciales en Python. La primera es hacerlo expl칤citamente:

````python
def sumar_cantidad(cantidad):
    def sumar(otro_elemento):
        return otro_elemento + cantidad
    
    return sumar

sumar_dos = sumar_cantidad(2) # creas una funcion aplicada parcialmente

print(f"2 + 4 = {sumar_dos(4)} | 10 + 2 = {sumar_dos(10)}")
````

Otra manera es usar el m칠todo `partial`:

````python
from functools import partial

def sumar(a, b):
    return a + b

sumar_dos = partial(sumar, b=2)  # hay que nombrar los parametros que vamos a "sobrescribir"

print(f"2 + 4 = {sumar_dos(4)} | 10 + 2 = {sumar_dos(10)}")
````
